## Joining the DataFrames.

```python
#note: on a un espace dans le nom de column 'listing_id' du reviews
reviews.columns = reviews.columns.str.strip()
```

```python
reviews.columns
```

```python
price_reviews = price.merge(reviews, how="outer", on='listing_id')
data = price_reviews.merge(room_type, how="outer", on='listing_id')
```

```python
data.head(2)
```

## le prix moyen des annonces Airbnb au marché locatif privé par mois.

**Objectif : Analyser comment les coûts moyens des annonces Airbnb se comparent au marché locatif traditionnel.**

le prix moyen journalier des annonces Airbnb

```python
avg = round(data['price'].mean(),2)
print('Le prix moyen est :', avg ,'par nuit')
```

 le coût mensuel moyen (multiplication par 30 jours).

```python
avg_month = round(data['price'].mean()*30,2)
print('Le prix moyen est :', avg_month ,'par mois')
```

Comparer ces données au coût moyen mensuel du marché locatif privé (ex. $3,100)

**Airbnb est généralement plus cher à long terme que le marché locatif privé.**

## la variation des prix des annonces Airbnb dans les cinq boroughs de New York.

```python
data['nbhood_full'].value_counts()
```

extraire information de la column nbhood_full a nouveau column

```python
data['nbhood'] = data['nbhood_full'].str.split(',').str[0]
```

```python
data['nbhood'].unique()
```

```python
# statistics
stat = data.groupby("nbhood")["price"].agg(["sum", "mean", "median", "count"])

stat = stat.round(2).sort_values("mean", ascending=False)
```
